1. Implement selection sort on an array of integers and print the array after each iteration to
show the sorting progress.

```python
%%writefile 1.c
#include <stdio.h>

void selectionSort(int arr[], int n) {
    int i, j, minIndex, temp;

    for (i = 0; i < n - 1; i++) {
        minIndex = i;


        for (j = i + 1; j < n; j++) {
            if (arr[j] < arr[minIndex]) {
                minIndex = j;
            }
        }


        temp = arr[i];
        arr[i] = arr[minIndex];
        arr[minIndex] = temp;

        printf("After iteration %d: ", i + 1);
        for (int k = 0; k < n; k++) {
            printf("%d ", arr[k]);
        }
        printf("\n");
    }
}

int main() {
    int arr[] = {64, 25, 12, 22, 11};
    int n = sizeof(arr) / sizeof(arr[0]);

    printf("Original array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n\n");

    selectionSort(arr, n);

    printf("\nSorted array: ");
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }

    return 0;
}
```

3. Design a function that performs partial selection sort â€” it only sorts the first k smallest
elements

```python
%%writefile 3.c

#include <stdio.h>

void partialSelectionSort(int arr[], int n, int k) {
    int i, j, minIndex, temp;


    for (i = 0; i < k; i++) {
        minIndex = i;


        for (j = i + 1; j < n; j++) {
            if (arr[j] < arr[minIndex]) {
                minIndex = j;
            }
        }

        temp = arr[i];
        arr[i] = arr[minIndex];
        arr[minIndex] = temp;
    }
}

int main() {
    int arr[] = {64, 25, 12, 22, 11, 90, 5};
    int n = sizeof(arr) / sizeof(arr[0]);
    int k = 3;

    partialSelectionSort(arr, n, k);

    printf("Array after sorting first %d smallest elements:\n", k);
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }

    return 0;
}
```

5. Write a function to merge two sorted linked lists into one sorted linked list. Do not use
arrays or extra data structures.

```python
%%writefile 5.c
#include <stdio.h>
#include <stdlib.h>


struct Node {
    int data;
    struct Node* next;
};


struct Node* newNode(int data) {
    struct Node* node = (struct Node*)malloc(sizeof(struct Node));
    node->data = data;
    node->next = NULL;
    return node;
}

// Merge two sorted linked lists without extra structures
struct Node* mergeSortedLists(struct Node* L1, struct Node* L2) {
    if (!L1) return L2;
    if (!L2) return L1;

    struct Node* head = NULL;
    struct Node* tail = NULL;

    if (L1->data < L2->data) {
        head = tail = L1;
        L1 = L1->next;
    } else {
        head = tail = L2;
        L2 = L2->next;
    }

    while (L1 && L2) {
        if (L1->data < L2->data) {
            tail->next = L1;
            L1 = L1->next;
        } else {
            tail->next = L2;
            L2 = L2->next;
        }
        tail = tail->next;
    }

    if (L1) tail->next = L1;
    else    tail->next = L2;

    return head;
}


void printList(struct Node* head) {
    while (head) {
        printf("%d ", head->data);
        head = head->next;
    }
    printf("\n");
}

int main() {
    // Creating list 1: 1 -> 3 -> 5
    struct Node* L1 = newNode(1);
    L1->next = newNode(3);
    L1->next->next = newNode(5);

    // Creating list 2: 2 -> 4 -> 6
    struct Node* L2 = newNode(2);
    L2->next = newNode(4);
    L2->next->next = newNode(6);

    printf("List 1: ");
    printList(L1);

    printf("List 2: ");
    printList(L2);


    struct Node* merged = mergeSortedLists(L1, L2);
    printf("Merged List: ");
    printList(merged);

    return 0;
}
```

6. Implement a merge procedure using two queues (each containing sorted integers) and
output a single sorted queue.

```python
%%writefile 6.c

#include <stdio.h>

#define MAX 100

struct Queue {
    int arr[MAX];
    int front;
    int rear;
};


void init(struct Queue *q) {
    q->front = 0;
    q->rear = -1;
}


int isEmpty(struct Queue *q) {
    return q->front > q->rear;
}


void enqueue(struct Queue *q, int value) {
    q->rear++;
    q->arr[q->rear] = value;
}


int dequeue(struct Queue *q) {
    int value = q->arr[q->front];
    q->front++;
    return value;
}


void merge(struct Queue *q1, struct Queue *q2, struct Queue *result) {

    while (!isEmpty(q1) && !isEmpty(q2)) {
        if (q1->arr[q1->front] < q2->arr[q2->front]) {
            enqueue(result, dequeue(q1));
        } else {
            enqueue(result, dequeue(q2));
        }
    }


    while (!isEmpty(q1)) {
        enqueue(result, dequeue(q1));
    }

    while (!isEmpty(q2)) {
        enqueue(result, dequeue(q2));
    }
}

int main() {

    struct Queue q1, q2, result;
    init(&q1);
    init(&q2);
    init(&result);

    // two arrays to create two queues
    int a1[] = {1, 4, 7, 10};
    int a2[] = {2, 3, 6, 9, 11};

    int n1 = 4, n2 = 5;

    // Filling queue 1
    for (int i = 0; i < n1; i++) {
        enqueue(&q1, a1[i]);
    }

    // Filling queue 2
    for (int i = 0; i < n2; i++) {
        enqueue(&q2, a2[i]);
    }


    merge(&q1, &q2, &result);


    printf("Merged Sorted Queue: ");
    while (!isEmpty(&result)) {
        printf("%d ", dequeue(&result));
    }

    return 0;
}
```

7. Write a hybrid program that divides an array into chunks, performs selection sort on each
chunk, and then merges the chunks using your merge procedure.

```python
%%writefile 7.c

#include <stdio.h>

void print_array(int arr[], int n) {
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n");
}


void selection_sort(int arr[], int l, int r) {
    for (int i = l; i <= r; i++) {
        int min_index = i;
        for (int j = i + 1; j <= r; j++) {
            if (arr[j] < arr[min_index]) {
                min_index = j;
            }
        }
        int temp = arr[i];
        arr[i] = arr[min_index];
        arr[min_index] = temp;
    }
}


void merge(int arr[], int l, int m, int r) {
    int temp[100];
    int i = l, j = m + 1, k = l;

    while (i <= m && j <= r) {
        if (arr[i] <= arr[j]) temp[k++] = arr[i++];
        else temp[k++] = arr[j++];
    }
    while (i <= m) temp[k++] = arr[i++];
    while (j <= r) temp[k++] = arr[j++];

    for (int x = l; x <= r; x++) {
        arr[x] = temp[x];
    }
}

int main() {
    int arr[] = {10, 3, 5, 30, 2, 8, 20, 15};
    int n = sizeof(arr) / sizeof(arr[0]);

    printf("Original array:\n");
    print_array(arr, n);

    int mid = (n - 1) / 2;


    selection_sort(arr, 0, mid);


    selection_sort(arr, mid + 1, n - 1);

    printf("\nAfter sorting two halves separately:\n");
    print_array(arr, n);

    merge(arr, 0, mid, n - 1);

    printf("\nFinal merged sorted array:\n");
    print_array(arr, n);

    return 0;
}
```
